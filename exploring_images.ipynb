# Let's look at some cat histograms!

```python
from os import listdir, getcwd
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


from histograms import plot_image, split_layers_to_3, plot_histogram, plot_histograms
```

<br><br>
## Read in image filenames


```python
animals_lst = listdir('images')

if '.DS_Store' in animals_lst:
    animals_lst.remove('.DS_Store')

cwd = getcwd()

animals_lst = [cwd + '/images/' + filename for filename in animals_lst]
```

<br><br>
## Plot a single image and a histogram of one of its channels


```python
# view a specific image
choice = animals_lst[4]

cat_image = imread(choice)

# print(type(cat_image))
# print(cat_image.shape)

plot_image(cat_image, choice)

r, g, b = split_layers_to_3(cat_image)

print(r.shape)
print(g.shape)
print(b.shape)
```

```python
plot_histogram(r, animals_lst[1])
```

<br><br>
## Plot histograms for all channels, luminosity and averaged channel
* note that the averaged channel is simply the avg of each channel value for each pixel, which is different from grayscale in terms of the weights given to each channel.

```python
plot_histograms(cat_image, choice, images=True)
```
